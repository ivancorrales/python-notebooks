## Variables en Python: tipado din√°mico

- En Python **no hace falta declarar** las variables antes de usarlas: **nacen** cuando les asignas un valor por primera vez.
- Python tiene **tipado din√°mico**: el **tipo** de una variable depende del **valor** que tenga en ese momento.
- Es **posible** (aunque **no recomendable**) **cambiar el tipo** de una variable reasign√°ndola.

### ¬øPor qu√© no conviene cambiar de tipo?
- **Legibilidad**: confunde a quien lee el c√≥digo.
- **Errores sutiles**: operaciones que funcionaban dejan de hacerlo.
- **Herramientas**: dificulta el autocompletado y el chequeo est√°tico (por ejemplo, con mypy).

````python
# Definici√≥n de variables
numero_entero = 42
numero_flotante = 3.14
es_valido = True
texto = "Hola, DAM"
lista = [1, 2, 3, "cuatro"]
conjunto = {1, 2, 3}
mapa = {"nombre": "Ana", "edad": 19}
nada = None

# Imprime primero solo el valor, y luego su tipo (sin for ni funciones)

print(numero_entero)
print(f"tipo: {type(numero_entero).__name__}\n")

print(numero_flotante)
print(f"tipo: {type(numero_flotante).__name__}\n")

print(es_valido)
print(f"tipo: {type(es_valido).__name__}\n")

print(texto)
print(f"tipo: {type(texto).__name__}\n")

print(lista)
print(f"tipo: {type(lista).__name__}\n")

print(conjunto)
print(f"tipo: {type(conjunto).__name__}\n")

print(mapa)
print(f"tipo: {type(mapa).__name__}\n")

print(nada)
print(f"tipo: {type(nada).__name__}")
````

````python
# La variable se "crea" al asignarla por primera vez
x = 10                     # int
print(x, type(x))

# Reasignaci√≥n a otro tipo (NO recomendado)
x = "diez"                 # str
print(x, type(x))

# Ejemplo de problema t√≠pico al mezclar tipos
total = 5                  # int
total += 2                 # 7
print("total:", total, type(total))

total = "siete"            # ahora es str
print("total:", total, type(total))

# Esto falla porque intentamos sumar un int a un str
try:
    total += 3
except TypeError as e:
    print("Error:", e)
````

üëâ Recomendaci√≥n: mant√©n el mismo tipo para cada nombre de variable. Si necesitas otro tipo, usa un nombre nuevo o convierte el valor expl√≠citamente (int(), float(), str(), etc.). Usa type(var) para inspeccionar el tipo en tiempo de ejecuci√≥n.

````python
# Mant√©n un tipo consistente por nombre de variable
precio_euros = 9.99        # float
print(precio_euros, type(precio_euros))

# Si necesitas otra representaci√≥n, usa otro nombre o convierte expl√≠citamente
precio_texto = str(precio_euros)   # str
print(precio_texto, type(precio_texto))

# Otro ejemplo con conversi√≥n expl√≠cita
edad_str = "18"            # str que viene de input/formulario
# edad_str += 1  # (esto fallar√≠a)
edad = int(edad_str)       # convertimos a int
print(edad, type(edad))


nota: float = 7.5
print(nota, type(nota))

# Cambiar el tipo aqu√≠ funcionar√≠a en tiempo de ejecuci√≥n, pero es mala pr√°ctica
# y un verificador est√°tico lo marcar√≠a:
# nota = "siete y medio"   # ‚ùå evita esto
````

## Operaciones

### Operaciones con n√∫meros enteros

````python
x = 3

print("- Tipo de x:")
print(type(x)) # Imprime el tipo (o `clase`) de x
print("- Valor de x:")
print(x)       # Imprimir un valor
print("- x+1:")
print(x + 1)   # Suma: imprime "4"
print("- x-1:")
print(x - 1)   # Resta; imprime "2"
print("- x*2:")
print(x * 2)   # Multiplicaci√≥n; imprime "6"
print("- x^2:")
print(x ** 2)  # Exponenciaci√≥n; imprime "9"
# Modificaci√≥n de x
x += 1
print("- x modificado:")
print(x)  # Imprime "4"

x *= 2
print("- x modificado:")
print(x)  # Imprime "8"

print("- Varias cosas en una l√≠nea:")
print(1,2,x,5*2) # imprime varias cosas a la vez
````

### Operaciones con n√∫meros decimales (float)

````python
y = 2.5
print("- Tipo de y:")
print(type(y)) # Imprime el tipo de y
print("- Varios valores en punto flotante:")
print(y, y + 1, y * 2.5, y ** 2) # Imprime varios n√∫meros en punto flotante
````

### Operaciones con Booleanos
Python implementa todos los operadores usuales de la l√≥gica booleana, usando palabras en ingl√©s (and, or, not) en lugar de s√≠mbolos (||, &&, !, etc)

Tambi√©n tiene los t√≠picos operadores de comparaci√≥n: <,>,>=,<=,==,!=

````python
# Operaciones con Booleanos y comparaciones en Python

# Booleanos b√°sicos
a = True
b = False

print("a AND b  ->", a and b)   # and
print("a OR b   ->", a or b)    # or
print("NOT a    ->", not a)     # not
print("NOT b    ->", not b)

# Cortocircuito (short-circuit):
# En 'and', si la izquierda es False, no eval√∫a la derecha.
# En 'or', si la izquierda es True, no eval√∫a la derecha.
print("False and (1/0) ->", False and (1/0))  # seguro: no eval√∫a 1/0
print("True  or  (1/0) ->", True  or  (1/0))  # seguro: no eval√∫a 1/0

# Operadores de comparaci√≥n
x = 3
y = 5

print("x < y   ->", x < y)
print("x > y   ->", x > y)
print("x >= 3  ->", x >= 3)
print("x <= 3  ->", x <= 3)
print("x == 3  ->", x == 3)
print("x != y  ->", x != y)

# Combinando comparaciones con booleanos
print("(x < y) and (y < 10) ->", (x < y) and (y < 10))

# Comparaci√≥n de cadenas de texto (lexicogr√°fica, por Unicode)
s1 = "Ana"
s2 = "Juan"
print('"Ana" < "Juan" ->', s1 < s2)
````

````python
print("- Lista con 4 n√∫meros:")
a=[20,15,7,13] # una lista con cuatro n√∫meros
print(a)

print("- Lista con 3 strings:")
b=["hola","adios","buen d√≠a"] # una lista con tres strings
print(b)

# la funci√≥n `len` me da la longitud de la lista
print("- Longitud de la lista:")
n=len(a)
print(n)
````

````python
# Para acceder a sus elementos, se utiliza el []
# Los √≠ndices comienzan en 0

# Lista de ejemplo
b = ["manzana", "pera", "uva", "pl√°tano"]

print("- Elemento con √≠ndice 0 de la lista:")
print(b[0])

print("- Elemento con √≠ndice 1 de la lista:")
print(b[1])

print("- Elemento con √≠ndice 2 de la lista:")
print(b[2])
````

````python
# Ejemplos de lista vac√≠a y slicing (sublistas)

# Lista de ejemplo
a = ["A", "B", "C", "D", "E"]

# Para crear una lista vac√≠a (sin elementos), simplemente ponemos []
vacia = []
print("Lista vac√≠a:")
print(vacia)

# Tambi√©n puedes crear una sub-lista (slice) especificando un rango de √≠ndices [inicio:fin)
# Nota: el √≠ndice inicial se incluye y el final se excluye

print("- Elementos del √≠ndice 0 al 1 (2-1):")
print(a[0:2])   # ["A", "B"]

print("- Elementos del √≠ndice 1 al 3 (4-1):")
print(a[1:4])   # ["B", "C", "D"]
````

**¬øQu√© mostrar√≠a en cada uno de los siguientes ejemplos?**

```python
# Lista de ejemplo
a = ["A", "B", "C", "D", "E"]

# Si no pones nada antes de ":" se asume 0
print("- Elementos desde el comienzo al √≠ndice 1 (2-1):")
print(a[:2])    # ["A", "B"]

# Si no pones nada despu√©s de ":" se asume hasta el final
print("- Elementos desde el √≠ndice 1 hasta el final:")
print(a[1:])    # ["B", "C", "D", "E"]

# Si no pones nada ni antes ni despu√©s es como tomar todo
print("- Todos los elementos:")
print(a[:])     # ["A", "B", "C", "D", "E"]

# Si el fin es igual al comienzo, es un rango vac√≠o ‚Üí lista vac√≠a
print("- Rango vac√≠o -> lista vac√≠a:")
print(a[2:2])   # []
```

## Estructuras de control
En Python no hay llaves ({}) ni begin...end para marcar el comienzo y fin de un bloque, sino que eso se logra con la indentaci√≥n. La indentaci√≥n por defecto son 4 espacios en blanco.

Entonces va a ser necesario indentar correctamente para utilizar sentencias if, for o para definir funciones.

if
El if es como el de otros lenguajes, pero no pide par√©ntesis y termina con :. Su sintaxis es:

```python
if condicion :
    cuerpo del if (indentado con 4 espacios)
else:
    cuerpo del else (indentado con 4 espacios)
```

````python
rol = "admin"

print("El usuario es")
if rol == "admin":  # el if termina con ':' para indicar el final de la condici√≥n
    # este print va indentado (4 espacios) porque est√° dentro del cuerpo del if
    print("Administrador")
else:               # 'else:' cubre el caso contrario
    # tambi√©n va indentado porque pertenece al bloque del else
    print("Est√°ndar")

print("del sistema")
````
